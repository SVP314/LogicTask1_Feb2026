```python
 # Problem 1: Smart Light Controller
class smartlight:
    def setlight(self, name, action):
        self.lightName = name
        self.status = action

    def display(self):
        print(self.lightName + " is " + self.status)

# taking input from user
name = input("Light Name: ")
action = input("Action: ")

# creating object and displaying
bedroomlight = smartlight()
bedroomlight.setlight(name, action)
bedroomlight.display()
```

```python
# Problem 2: Employee ID Card System
class employeecard:
    def setdata(self, empname, empid, department):
        self.name = empname
        self.id = empid
        self.dept = department

    def show(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.id)
        print("Department:", self.dept)

# taking employee details from user
name = input("Employee Name: ")
id = input("Employee ID: ")
dept = input("Department: ")

# creating and showing id card
emp = employeecard()
emp.setdata(name, id, dept)
emp.show()
```

```python
# Problem 3: Mobile Contact Record
class mobilecontact:
    def save(self, contactname, phoneno):
        self.name = contactname
        self.number = phoneno

    def showcontact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.number)

# getting contact details from user
name = input("Contact Name: ")
phone = input("Phone Number: ")

# saving and displaying contact
c = mobilecontact()
c.save(name, phone)
c.showcontact()
```

```python
# Problem 4: Product Price Tag Generator
class product:
    def set(self, productname, p):
        self.pname = productname
        self.price = p

    def print(self):
        print("Product:", self.pname)
        print("Price: ₹" + str(self.price))

# taking product details from user
name = input("Product Name: ")
price = int(input("Price: "))  # converting to int for price

# generating price tag
p = product()
p.set(name, price)
p.print()
```

```python
# Problem 5: Movie Rating Display System
class movierating:
    def store(self, movie, rating):
        self.name = movie
        self.rating = rating

    def show(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/5")

# getting movie info from user
name = input("Movie Name: ")
rating = float(input("Rating: "))  # converting to float for rating

# storing and displaying movie
m = movierating()
m.store(name, rating)
m.show()
```

```python
# Problem 6: Delivery Address Manager
class deliveryaddress:
    def details(self, name, address):
        self.customer = name
        self.address = address

    def print(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)

# taking delivery details from user
name = input("Customer Name: ")
address = input("Address: ")

# saving and printing delivery info
d = deliveryaddress()
d.details(name, address)
d.print()
```
