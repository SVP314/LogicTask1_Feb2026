```python
# problem statement 1: smart parking lot management system
def parking_system(capacity, logs):
    # calculate total parked vehicles
    parked = 0
    for action in logs:
        if action == "IN":
            parked = parked + 1
        elif action == "OUT":
            parked = parked - 1

    # check if parking exceeds capacity
    if parked <= capacity:
        status = "Available"
    else:
        status = "Full"

    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)

# test case
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_system(50, logs)
```

```python
# problem statement 2: online food delivery time estimator
def delivery_estimator():
    # get user input exactly as shown in sample
    distance = float(input("Distance (km): "))
    traffic = input("Traffic Level: ")
    weather = input("Weather: ")

    # base speed
    if traffic == "Low":
        speed = 30  # km per hour
    elif traffic == "Moderate":
        speed = 20
    elif traffic == "High":
        speed = 15

    # weather impact on speed
    if weather == "Rainy":
        speed = speed - 5
    elif weather == "Stormy":
        speed = speed - 10
    elif weather == "Snowy":
        speed = speed - 15

    # convert to minutes
    time = (distance / speed) * 60

    print("Estimated Delivery Time:", int(time), "minutes")

# run
delivery_estimator()
```

```python
# problem statement 3: movie theatre seat occupancy analyzer
def theater_analyzer(total, booked):
    # calculate occupancy percentage
    count = 0
    for seat in booked:
        if seat == 1:
            count = count + 1

    percent = (count / total) * 100

    # determine if show is housefull
    if percent >= 90:
       status = "Housefull"
    # suggest opening additional shows
    elif percent >= 70:
         status = "Almost Full"
    # suggest opening additional shows
    else:
        status = "Not Full"
    # no additional shows needed

    # display output exactly as expected
    print("Occupancy:", round(percent), "%")
    print("Show Status:", status)

# test case
total_seats = 200
# creating 150 entries of 1
booked_seats = []
for i in range(150):
    booked_seats.append(1)

theater_analyzer(total_seats, booked_seats)
```

```python
# problem statement 4: cloud server load classification system
def server_load(readings):
    # calculate average cpu usage
    total = 0
    for i in readings:
        total = total + i

    average = total / len(readings)

    # classify based on rules
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # display output
    print("Average CPU Load:", round(average), "%")
    print("Server Status:", status)

# test case
cpu_readings = [45, 60, 70, 85, 90]
server_load(cpu_readings)
```

```python
# problem statement 5: smart classroom resource usage monitor
def classroom_monitor(usage):
    # find which resources are used too much
    overused = []

    # check each resource one by one
    for resource, hours in usage.items():
        # if used more than 8 hours:overused
        if hours > 8:
            overused.append(resource)

    # decide if energy alert is needed
    if len(overused) > 0:
        alert = "Yes"
        # combine multiple resources into one string
        if len(overused) == 1:
            over_text = overused[0]
        else:
            over_text = ""
            for r in overused:
                if over_text == "":
                    over_text = r
                else:
                    over_text = over_text + ", " + r
    else:
        alert = "No"
        over_text = "None"

    # show results
    print("Overused Resources:", over_text)
    print("Energy Alert:", alert)

# test
resource_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
classroom_monitor(resource_data)
```

```python
# problem statement 6: online event registration capacity controller
def event_controller(capacity, registrations):
    # check if registrations exceed capacity
    if registrations <= capacity:
        # still within limit, no waitlist needed
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        # over capacity, put extra people on waitlist
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    # results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

# test
event_controller(100, 105)
```
